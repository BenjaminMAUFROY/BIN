```python
from skimage import io, exposure
import matplotlib.pyplot as plt

# Load the input image 
image = io.imread("BN_Tri_entrainement/G0178831.JPG")

# Increase contrast by stretching the intensity range
high_contrast_image = exposure.rescale_intensity(image[:,:,0], in_range=(0, 180), out_range=(0, 1))

# Display the original and high-contrast images
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
ax1, ax2 = axes.ravel()

ax1.imshow(image[:,:,0], cmap='gray')
ax1.set_title('Original Image')
ax1.set_axis_off()

ax2.imshow(high_contrast_image, cmap='gray')
ax2.set_title('High Contrast Image')
ax2.set_axis_off()


plt.show()

# image_laplacian = filters.laplace(high_contrast_image)
# io.imshow(image_laplacian)
```

```python
# Data augmentation
from skimage import transform,data,filters, io, img_as_float
import matplotlib.pyplot as plt
import random as rd

image = io.imread("BN_Tri_entrainement/G0178831.JPG")

transformed_img = transform.rotate(image,
                                   angle = rd.randint(1,359),
                                   cval = 255
                                   )

fig, axes = plt.subplots(1, 2, figsize=(12, 6))
ax1, ax2 = axes.ravel()

ax1.imshow(image, cmap='gray')
ax1.set_title('Original Image')
ax1.set_axis_off()

ax2.imshow(transformed_img, cmap='gray')
ax2.set_title('rotate')
ax2.set_axis_off()

plt.show()
```
